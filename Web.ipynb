```python
title_list=[]
```

```python
rating_list=[]
```

```python
for page_num in range(1,51):
    reg_page = requests.get('https://books.toscrape.com/catalogue/page-2.html')
    soup_page = bs4.BeautifulSoup(reg_page.text)
    
    for item in soup_page.select('article h3 a'):
        title_list.append(item['title'])
        
    for n in range(0,60,3):
        rating_list.append(soup_page.select('article p')[n]['class'][1])
    
```
